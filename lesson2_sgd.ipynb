```python
import torch
from fastai.basics import *
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
n = 200
x = torch.ones(n, 2)
```

```python
x[:, 0].uniform_(-2, 2)
x[:5]
```

```python
def mse(y_hat, y):
    return ((y_hat - y) ** 2).mean()
```

```python
def relu(x):
    return x if x >= 0 else 0
```

```python
w = tensor(1.5, 3)
y = x ** 2 @ w + torch.rand(n)
```

```python
plt.figure()
plt.scatter(x[:, 0], y);
```

```python
a = tensor(1., 1)
a = nn.Parameter(a)
a
```

```python
def update():
    y_hat = x ** 2 @ a
    loss = mse(y_hat, y)
    if t % 10 == 0: print(loss)
    loss.backward()
    with torch.no_grad():
        a.sub_(1e-1 * a.grad)
        a.grad.zero_()
```

```python
for t in range(200): update()
```

```python
plt.figure()
plt.scatter(x[:, 0], y, label='true');
plt.scatter(x[:, 0], x** 2 @ a, label='pred')
plt.legend();
```
